# Witness Parser Workflow

This notebook provides a workflow for using the functions in the `witness_parser` module.

## Setup

Start by importing the Python modules and configuring the paths to your `front.xml` and `textualnotes.xml` files.

```python
# Python imports
import json
from witness_parser import get_witness_list, parse_notes, parse_notes_to_xml

# Configuration (edit these paths as needed)
NOTES_XML_PATH = "sample-data/mnd_textualnotes.xml"
FRONT_XML_PATH = "sample-data/mnd_front.xml"
```

## Get a List of all Front Matter Witnesses

By default, the cell below prints only the first 10 witnesses.

```python
witness_list = get_witness_list(FRONT_XML_PATH)
witness_list[0:10]
```

## Parse All the Notes

The cell below generates a dictionary of each XML textual note with a list of witnesses associated with each reading. By default, only the first ten notes are shown.

Note that an asterisk (*) indicates that the witnesses is listed in a secondary witness list in the front matter. Witnesses are ordered by date based on the order in which they are listed in the front matter. A rubric is used for ordering dates containing brackets, ranges, question marks, and other notations when combining primary and secondary witnesses. However, the order should be double checked to ensure the relative chronology is expected.

```python
# Generate the dictionary of notes and their readings
notes = parse_notes(NOTES_XML_PATH, FRONT_XML_PATH)

# Print the notes and their readings for inspection
for note in notes[0:10]:
    print(note["xml:id"], note.get("lem"))
    for reading in note["readings"]:
        print("  ", reading["rdg"], "→", reading.get("wit"))
```

## Save to JSON File

This cell will save the output to a JSON file in the desired location.

```python
with open("sample-data/textual_notes.json", "w", encoding="utf-8") as file:
    json.dump(notes, file, indent=4)
```

## Transform the Original XML

You can rewrite the original xml with the complete list of witnesses in a number of ways with the `parse_notes_to_xml` function (which returns a string). By default, this function rewrites the notes so that every `appPart` element has a `witnesses` attribute that references the `xml:id` of each witness containing the reading. By default, the original `wit` elements are retained for comparison. You can remove them by setting `remove_wit=True` or replace them with new `wit` elements containing the updated list of sigla by setting `replace_wit=True`.

Note: There is currently no way to remove the `witnesses` attribute from `appPart`, as that is considered essential for comparison during the experimental phase.

```python
xml_string = parse_notes_to_xml("sample-data/mnd_textualnotes.xml", "sample-data/mnd_front.xml", remove_wit=True)
print(xml_string[0:1000])
```

## Save to XML File

This cell will save the output to an XML file in the desired location.

```python
with open("sample-data/experimental_textual_notes.xml", "w", encoding="utf-8") as file:
    file.write(xml_string)
```
